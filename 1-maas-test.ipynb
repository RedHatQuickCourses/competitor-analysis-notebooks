This simple notebook tests if a query can be sent to the remote inference end point running on Models as a Service (MaaS)

```python
import requests
import urllib3
import json
import os
from pathlib import Path

API_URL = "https://granite-3-3-8b-instruct-maas-apicast-production.apps.prod.rhoai.rh-aiservices-bu.com:443"

API_KEY = os.environ.get("VLLM_API_TOKEN")

input = "What is the capital of Mongolia?"

completion = requests.post(
    url=API_URL+'/v1/completions',
    json={
      "model": "granite-3-3-8b-instruct",
      "prompt": input,
      "max_tokens": 15,
      "temperature": 0.2
    },
    headers={'Authorization': 'Bearer '+API_KEY}
).json()

if 'choices' in completion and len(completion['choices']) > 0:
    print(completion['choices'][0]['text'].strip())
else:
    print("Error: Could not find 'choices' in the API response.")
    print(completion)
```
